```python
import matplotlib.pyplot as plt
```

```python
#create a black 2D image (grayscale image)
black_img = np.zeros((200,200), dtype=np.uint8)

plt.imshow(black_img, cmap=plt.cm.gray) #define the map to grayscale
plt.title("black grayscale")
plt.xticks([]), plt.yticks([]) #remove the x and y axies
plt.show()
```

```python
# create a white image
white_img = 255 * np.ones((200,200))
white_img = np.uint8(white_img)

plt.imshow(white_img, cmap = plt.cm.gray, vmin = 0, vmax =256)
plt.title("White grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a random color image
random_color_img = np.random.randint(0, high = 256, size = (200,200,3), dtype = "uint8")

plt.imshow(random_color_img)
plt.title("random color")
plt.xticks([]), plt.yticks([])
plt.show()
```

## Exercise 1

Try to write a simple custom function to determine whether a given integer is add or even number.

```python
## 1. Try to write a simple custom function to 
## determine whether a given integer is add or even number.

def OddEven(x):     #for integers only
    """
    This function is to determine the odd even nature of number
    """
    if isinstance(x,int):
        if x%2 == 0:
            if x==0:
                statement = " is neither odd or even number."
            else :
                statement = " is even number."
        else:
            statement = " is odd number."
    else:
        statement = " is not an integer!"
        
        
    return print(f"{x}" + statement)
           
```

```python
OddEven(0)
```

```python
OddEven(0.9)
```

```python
OddEven(12)
```

# Exercise 4
Create an image as shown as the following with the help of Numpy and matplotlib modules. You can
arbitrarily set the dimension of the image and white circular spot at the middle.

```python
from matplotlib import pyplot as patches
```

```python
# create a black image
black_img = np.zeros((200, 200), dtype = "uint8")

#create a circle which have a radius of 30 and the color white
#plot the circle at the coordinates at (100,100) which is the have of the square axis
white_spot = plt.Circle((100,100),radius = 30, color = "white")

#add circle to plot
# gca means "get current axis"
plt.gca().add_artist(white_spot)

plt.imshow(black_img,cmap = plt.cm.gray)
plt.title("White Spot image")
plt.xticks([]), plt.yticks([])
plt.show()
```
